```python
#Создайте «признак-мигалку» weekend, который равен 1, если поездка начиналась в выходной день (суббота или воскресенье),
# и 0 — в противном случае. Выясните, сколько поездок начиналось в выходные.
# Чтобы создать признак weekend, необходимо с помощью атрибута аксессора dt dayofweek выделить день недели из столбца с датой начала поездки.
# С помощью метода apply() можно применить к результирующей Series функцию, которая принимает на вход число (день поездки) и возвращает 1, если это число равно 5 или 6 (суббота и воскресенье), в противном случае функция возвращает 0. Результат необходимо занести в столбец weekend.
# Самый простой способ посчитать число поездок, совершаемых в выходные — рассчитать сумму по столбцу (поскольку столбец состоит из 1 и 0, на выходе суммы получается количество, равное 1).
    
weekend = cb_df['starttime'].dt.day_of_week
cb_df['weekend'] = weekend.apply(lambda x: 1 if x == 5 or x == 6 else 0)
cb_df['weekend'].sum()
```

```python
# Создайте признак времени суток поездки time_of_day. Время суток будем определять из часа начала поездки. Условимся, что:

#     поездка совершается ночью (night), если её час приходится на интервал от 0 (включительно) до 6 (включительно) часов;
#     поездка совершается утром (morning), если её час приходится на интервал от 6 (не включительно) до 12 (включительно) часов;
#     поездка совершается днём (day), если её час приходится на интервал от 12 (не включительно) до 18 (включительно) часов;
#     поездка совершается вечером (evening), если её час приходится на интервал от 18 (не включительно) до 23 часов (включительно).

# Во сколько раз количество поездок, совершённых днём, больше, чем количество поездок, совёршенных ночью,
# за представленный в данных период времени? Ответ округлите до целых.

    # Подсказка (1 из 3): Определите функцию get_time_of_day, которая принимает на вход время (в часах) 
    # и возвращает время суток в зависимости от принадлежности часа к указанному в задании диапазону.
    # Подсказка (2 из 3): Примените эту функцию к столбцу starttime, предварительно выделив час начала поездки
    # с помощью атрибута аксессора dt hour. Результат занесите в столбец time_of_day.
    # Подсказка (3 из 3): С помощью фильтрации найдите число поездок, совершённых днем и ночью, и разделите первое на второе.
def get_time_of_day(time):
    if 0 <= time <= 6:
        return 'night'
    elif 6 < time <= 12:
        return 'morning'
    elif 12 < time <= 18:
        return 'day'
    elif 18 < time <= 23:
        return 'evening'
    
cb_df['time_of_day'] = cb_df['starttime'].dt.hour.apply(get_time_of_day)
a = cb_df[cb_df['time_of_day'] == 'day'].shape[0]
b = cb_df[cb_df['time_of_day'] == 'night'].shape[0]
print(round(a/b))
```
